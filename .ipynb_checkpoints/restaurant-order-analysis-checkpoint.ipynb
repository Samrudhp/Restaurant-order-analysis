```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data into csv
order_details = pd.read_csv('order_details.csv')
menu = pd.read_csv('menu_items.csv')
```

```python
#checking the columns
print(order_details.columns)
print(menu.columns)
```

```python
data = pd.merge(order_details, menu, left_on='item_id', right_on='menu_item_id')
```

```python
print(data.info())
```

```python
dish_counts = data['item_name'].value_counts()
print(dish_counts)
```

```python
plt.figure(figsize = (10,6))
dish_counts[:10].plot(kind = 'bar')
plt.title('Top 10 Most Ordered Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Order count')
plt.xticks(rotation = 45)
plt.show()
```

```python
#calculate total sales amount for each dish

data['total_sales'] = data['price']
dish_sales = data.groupby('item_name')['total_sales'].sum().sort_values(ascending=False)
print(dish_sales)
```

```python
plt.figure(figsize=(10,6))
dish_sales[:10].plot(kind = 'bar')
plt.title('top 10 dishes by sales amount')
plt.xlabel('Dish name')
plt.ylabel('Sales amount')
plt.show()
```

```python
data['order_time'] = pd.to_datetime(data['order_time'])
data['hour'] = data['order_time'].dt.hour
order_counts_by_hour = data['hour'].value_counts().sort_index()
print(order_counts_by_hour)
```

```python
plt.figure(figsize=(10,6))
order_counts_by_hour.plot(kind = 'line',marker = 'o')
plt.title('Order counts by hour of day')
plt.xlabel('Hour of day')
plt.ylabel('Order Count')
plt.grid(True)
plt.show()
```
